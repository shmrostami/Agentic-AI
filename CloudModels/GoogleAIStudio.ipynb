```python
import os
from IPython.display import display, Markdown

import google.genai as genai

client = genai.Client(api_key=os.getenv("GOOGLEAISTUDIO_API_KEY"))

response = client.models.generate_content(
    model='models/gemini-2.5-flash',          # no "models/" prefix
    contents="What is the capital of Iran?"
)

display(Markdown(response.text))
```

```python
for model in client.models.list():
    print(model.name)
```
